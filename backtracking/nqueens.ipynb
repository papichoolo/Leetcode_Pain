

There are rows, columns, positive and negative diagonals. We iterate through the rows making sure we dont pass the same cols, pos and neg diagonals

```python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        board= [['.']* n for _ in range(n)]
        
        col=set()
        posdiag=set()
        negdiag=set()
        res=[]
        def backtrack(i):
            if i==n:
                copy=["".join(bruh) for bruh in board]
                res.append(copy)   
                return
            #i=row, j= cols
            
            for j in range(n):
                if j in col or (i+j) in posdiag or (i-j) in negdiag:
                    continue
                col.add(j)
                posdiag.add(i+j)
                negdiag.add(i-j)
                board[i][j]="Q"

                backtrack(i+1)
                col.remove(j)
                posdiag.remove(i+j)
                negdiag.remove(i-j)
                board[i][j]="."
        backtrack(0)
        return res
            

        
```
