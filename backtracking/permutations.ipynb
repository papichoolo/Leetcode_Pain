```python
class Solution:
    def permute(self, nums):
        res=[]

        if len(nums)==0:
            return [[]]
        
        perms=self.permute(nums[1:])
        for p in perms:
            for i in range(len(p)+1):
                p_=p.copy()
                p_.insert(i,nums[0])
                res.append(p_)
        
        return res
        

bruh=Solution()
x=bruh.permute('ab')
x= [''.join(i) for i in x]
print(x)

flag=False
for i in x:
    if i in "baboey":
        flag=True
        break
print(flag)
```

```python
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        if len(s1) > len(s2):
            return False
        s1count, s2count = {}, {}
        for i in range(len(s1)):
            s1count[s1[i]] = 1 + s1count.get(s1[i], 0)
            s2count[s2[i]] = 1 + s2count.get(s2[i], 0)
        matches = 0

        for char in s1count:
            if char in s2count and s1count[char] == s2count[char]:
                matches += 1

        l = 0
        for r in range(len(s1), len(s2)):
            if matches == len(s1count):
                return True

            char_r = s2[r]
            s2count[char_r] = 1 + s2count.get(char_r, 0)
            if char_r in s1count and s2count[char_r] == s1count[char_r]:
                matches += 1
            elif char_r in s1count and s2count[char_r] == s1count[char_r] + 1:
                matches -= 1

            char_l = s2[l]
            s2count[char_l] -= 1
            if char_l in s1count and s2count[char_l] == s1count[char_l]:
                matches += 1
            elif char_l in s1count and s2count[char_l] == s1count[char_l] - 1:
                matches -= 1

            l += 1

        return matches == len(s1count)

bruh = Solution()
print(bruh.checkInclusion('ab', 'eidbaooo'))
```
